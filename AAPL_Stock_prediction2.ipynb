```python
from statsmodels.tsa.stattools import adfuller
def test_stationarity(timeseries):
    
    #Determing rolling statistics
    rolmean = timeseries.rolling(12).mean()
    rolstd = timeseries.rolling(12).std()

    #Plot rolling statistics:
    orig = plt.plot(timeseries, color='blue',label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean', linewidth=1)
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show(block=False)
    
    #print('Results of Dickey-Fuller Test:')
    #dftest = adfuller(timeseries, autolag='AIC')
    #dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    #for key,value in dftest[4].items():
    #    dfoutput['Critical Value (%s)'%key] = value
    #print(dfoutput,'%f' % (1/10**8))
 
```

```python
test_stationarity(df.Open)
#the variation in standard deviation is small but mean is increasing with time so we conclude that this is not a stationary series
```

```python
dfOpen_log = np.log(df.Open)
plt.plot(dfOpen_log)
```

```python
moving_avg = dfOpen_log.rolling(12).mean()
plt.plot(dfOpen_log)
plt.plot(moving_avg, color='red', linewidth=0.5)
```

```python
dfOpen_log_moving_avg_diff = dfOpen_log - moving_avg
dfOpen_log_moving_avg_diff.head(12)
```

```python
dfOpen_log_moving_avg_diff.dropna(inplace=True)
test_stationarity(dfOpen_log_moving_avg_diff)
```

```python
dfOpen_log_diff = dfOpen_log - dfOpen_log.shift()
plt.plot(dfOpen_log_diff)
```

```python
dfOpen_log_diff.dropna(inplace=True)
test_stationarity(dfOpen_log_diff)
```

```python
#ACF and PACF plots:
from statsmodels.tsa.stattools import acf, pacf
```

```python
lag_acf = acf(dfOpen_log_diff["2017":"2018"], nlags=40)
lag_pacf = pacf(dfOpen_log_diff["2017":"2018"], nlags=40, method='ols')
```

```python
#Plot ACF: 
plt.subplot(121) 
plt.plot(lag_acf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(dfOpen_log_diff["2017":"2018"])),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(dfOpen_log_diff["2017":"2018"])),linestyle='--',color='gray')
plt.title('Autocorrelation Function')


#Plot PACF:
plt.subplot(122)
plt.plot(lag_pacf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(dfOpen_log_diff["2017":"2018"])),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(dfOpen_log_diff["2017":"2018"])),linestyle='--',color='gray')
plt.title('Partial Autocorrelation Function')
plt.tight_layout()
```

```python
from statsmodels.tsa.arima_model import ARIMA
```

```python
model = ARIMA(dfOpen_log["2017":"2018"], order=(25, 1, 2))  
results_ARIMA = model.fit(disp=-1)  
plt.plot(dfOpen_log_diff["2017":"2018"])
plt.plot(results_ARIMA.fittedvalues, color='red')
plt.title('RSS: %.4f'% sum((results_ARIMA.fittedvalues-dfOpen_log_diff)**2))
```

```python
predictions_ARIMA_diff = pd.Series(results_ARIMA.fittedvalues, copy=True)
print (predictions_ARIMA_diff.head())
```

```python
predictions_ARIMA_diff_cumsum = predictions_ARIMA_diff.cumsum()
print (predictions_ARIMA_diff_cumsum.head())
```

```python
predictions_ARIMA_log = pd.Series(dfOpen_log["2017":"2018"].iloc[0], index=dfOpen_log["2017":"2018"].index)
predictions_ARIMA_log = predictions_ARIMA_log.add(predictions_ARIMA_diff_cumsum,fill_value=0)
predictions_ARIMA_log.head()
```

```python
predictions_ARIMA = np.exp(predictions_ARIMA_log)
plt.plot(df.Open["2017":"2018"])
plt.plot(predictions_ARIMA["2017":"2018"])
plt.title('RMSE: %.4f'% np.sqrt(sum((predictions_ARIMA-df.Open)**2)/len(df.Open)))
```
